# Graph Transformer Baseline — Month-1 GDP Nowcasting

**Benchmark model for comparison with V16 GNN, ARIMAX, XGBoost, and LSTM**

Uses the same data sources, same walk-forward protocol, same month-1 cutoff,
and same SAAR evaluation as the V16 GNN notebook. The only difference is the
model: a **Graph Transformer** replaces the Graph Neural Network.

### Design

| Parameter | Value |
|-----------|-------|
| **Model** | Graph Transformer (PyTorch) |
| **Target** | Quarterly GDP level |
| **Features** | Top 5 by \|corr\| with GDP (per-fold) + lagged GDP |
| **Data cutoff** | Month-1 (Jan/Apr/Jul/Oct) — ~45 day lead |
| **Graph** | Correlation-based edges (threshold=0.5) |
| **Hyperparameters** | Fixed — no tuning or grid search |
| **Walk-forward** | 80 folds (2006-Q1 to 2025-Q4) |
| **Evaluation** | SAAR MAE (same as V16 GNN) |

### Why Graph Transformer as baseline?

- Combines **graph structure** (correlation between indicators) with
  **self-attention** (learned feature weighting)
- Graph edges encode known relationships; attention learns latent ones
- More expressive than standard GNN message passing
- Uses same month-1 data cutoff → apples-to-apples timing comparison
- Per-fold feature selection → no data leakage
- Fixed hyperparameters — no tuning, no grid search

## Step 3 — Data Loading

Same loaders as V16 GNN — FRED monthly, MRTS monthly, GDP quarterly.

```python
def load_fred_monthly() -> pd.DataFrame:
    """Load all monthly FRED CSVs."""
    log("=" * 65)
    log("LOADING FRED MONTHLY DATA")
    log("=" * 65)

    dfs = []
    for fpath in sorted(FRED_DIR.glob("Monthly_*.csv")):
        try:
            df = pd.read_csv(fpath, parse_dates=["observation_date"])
            val_col = [c for c in df.columns if c != "observation_date"][0]
            df.rename(columns={"observation_date": "date"}, inplace=True)
            df[val_col] = pd.to_numeric(df[val_col], errors="coerce")
            df = df.dropna(subset=[val_col])
            df = df.set_index("date")[[val_col]]

            stem = fpath.stem.replace("Monthly_", "")
            parts = stem.split("_")
            col_name = (f"F_{parts[0]}_{parts[1][:4]}"
                        if len(parts) >= 2 else f"F_{parts[0]}")
            df.columns = [col_name]

            dfs.append(df)
            first = df.index.min().strftime("%Y-%m")
            last  = df.index.max().strftime("%Y-%m")
            log(f"  \u2713 {col_name:30s}  {len(df):4d} obs  "
                f"({first} \u2192 {last})")
        except Exception as e:
            log(f"  \u2717 {fpath.name}: {e}")

    if not dfs:
        raise RuntimeError("No FRED data loaded")

    merged = pd.concat(dfs, axis=1).sort_index()
    merged = merged.resample("MS").last()
    log(f"\n  FRED total: {merged.shape[1]} series, "
        f"{merged.shape[0]} months")
    return merged

fred_df = load_fred_monthly()
```

```python
def load_mrts_monthly() -> pd.DataFrame:
    """Load monthly MRTS data from Excel files."""
    log("\n" + "=" * 65)
    log("LOADING MRTS MONTHLY DATA (Excel, 1992-2026)")
    log("=" * 65)

    dfs = []
    for fpath in sorted(MRTS_DIR.glob("*.xlsx")):
        try:
            naics = fpath.name.split("_")[0]
            col_name = f"MRTS_{naics}"

            raw = pd.read_excel(fpath, header=None, skiprows=7)
            raw.columns = (["Period", "Value"]
                           + [f"extra_{i}"
                              for i in range(len(raw.columns) - 2)])
            raw = raw[["Period", "Value"]].copy()
            raw["Value"] = pd.to_numeric(raw["Value"], errors="coerce")
            raw = raw.dropna(subset=["Value"])
            raw = raw[raw["Period"].notna() & (raw["Period"] != "")]

            raw["date"] = pd.to_datetime(
                raw["Period"].astype(str), format="%b-%Y", errors="coerce")
            raw = raw.dropna(subset=["date"])
            raw = raw.set_index("date")[["Value"]]
            raw.columns = [col_name]

            dfs.append(raw)
            first = raw.index.min().strftime("%Y-%m")
            last  = raw.index.max().strftime("%Y-%m")
            log(f"  \u2713 {col_name:15s}  {len(raw):4d} obs  "
                f"({first} \u2192 {last})")
        except Exception as e:
            log(f"  \u2717 {fpath.name}: {e}")

    if not dfs:
        log("  \u26a0 No MRTS data loaded")
        return pd.DataFrame()

    merged = pd.concat(dfs, axis=1).sort_index()
    merged = merged.resample("MS").last()
    log(f"\n  MRTS total: {merged.shape[1]} series, "
        f"{merged.shape[0]} months")
    return merged

mrts_df = load_mrts_monthly()
```

```python
def load_gdp_quarterly() -> pd.DataFrame:
    """Load quarterly GDP data."""
    log("\n" + "=" * 65)
    log("LOADING GDP QUARTERLY DATA")
    log("=" * 65)

    df = pd.read_csv(GDP_CSV, parse_dates=["observation_date"])
    df.rename(columns={"observation_date": "date"}, inplace=True)
    df["GDP"] = pd.to_numeric(df["GDP"], errors="coerce")
    df = df.dropna(subset=["GDP"]).set_index("date")[["GDP"]]

    log(f"  {len(df)} quarters  "
        f"({df.index.min().strftime('%Y-Q')} "
        f"\u2192 {df.index.max().strftime('%Y-Q')})")
    return df

gdp_q = load_gdp_quarterly()
```

## Step 4 — Merge Data

Combine FRED + MRTS + GDP into a single monthly DataFrame.
Feature selection is done per-fold in the walk-forward loop (leakage-free).

```python
def merge_all_data(fred_df, mrts_df, gdp_q):
    monthly_idx = pd.date_range(
        start=min(fred_df.index.min(),
                  mrts_df.index.min() if len(mrts_df)
                  else fred_df.index.min()),
        end=max(fred_df.index.max(),
                mrts_df.index.max() if len(mrts_df)
                else fred_df.index.max()),
        freq="MS",
    )
    gdp_monthly = gdp_q.reindex(monthly_idx).ffill()

    parts = [fred_df]
    if len(mrts_df) > 0:
        parts.append(mrts_df)
    parts.append(gdp_monthly)
    merged = pd.concat(parts, axis=1).sort_index()

    # Handle duplicate columns
    if merged.columns.duplicated().any():
        new_cols, seen = [], {}
        for c in merged.columns:
            if c in seen:
                seen[c] += 1
                new_cols.append(f"{c}_{seen[c]}")
            else:
                seen[c] = 0
                new_cols.append(c)
        merged.columns = new_cols

    all_cols = list(merged.columns)
    log("\n" + "=" * 65)
    log("MERGED MONTHLY DATASET")
    log("=" * 65)
    log(f"  Shape: {merged.shape}")
    log(f"  Date range: {merged.index.min().strftime('%Y-%m')} "
        f"\u2192 {merged.index.max().strftime('%Y-%m')}")
    log(f"  Features: {len(all_cols) - 1}  (+GDP target)")
    return merged, all_cols

monthly_df, all_cols = merge_all_data(fred_df, mrts_df, gdp_q)
```

## Step 5 — Quarterly Aggregation with Month-1 Cutoff

For each quarter, take the month-1 value (Jan/Apr/Jul/Oct) for all features.
GDP is **lagged by 1 quarter** to avoid leaking the target.

```python
def build_quarterly_dataset(monthly_df: pd.DataFrame,
                            gdp_q: pd.DataFrame) -> pd.DataFrame:
    """Convert monthly data to quarterly using month-1 values only.

    For each quarter, take the value from month-1:
      Q1 -> January, Q2 -> April, Q3 -> July, Q4 -> October
    """
    month1_months = [1, 4, 7, 10]
    month1_data = monthly_df[
        monthly_df.index.month.isin(month1_months)].copy()

    # Map to quarter-start dates
    month1_data.index = month1_data.index.to_period("Q").to_timestamp()

    # Use PREVIOUS quarter's GDP as feature (avoid target leakage)
    if "GDP" in month1_data.columns:
        month1_data["GDP_feature"] = month1_data["GDP"].shift(1)
        month1_data = month1_data.drop(columns=["GDP"])

    # Add actual GDP as target
    month1_data = month1_data.join(gdp_q, how="left")

    log("\n" + "=" * 65)
    log("QUARTERLY DATASET (Month-1 Cutoff)")
    log("=" * 65)
    log(f"  Shape: {month1_data.shape}")
    log(f"  Date range: {month1_data.index.min().strftime('%Y-Q%q')} "
        f"\u2192 {month1_data.index.max().strftime('%Y-Q%q')}")
    log(f"  Features: {month1_data.shape[1] - 1}  "
        f"(GDP_feature = lagged GDP)")

    return month1_data

quarterly_df = build_quarterly_dataset(monthly_df, gdp_q)
print(f"\nFirst 5 rows:")
print(quarterly_df[["GDP", "GDP_feature"]].head())
```

## Step 6 — Graph Transformer Model Architecture

The Graph Transformer combines **graph structure** with **self-attention**:

1. **Graph construction**: Edges between features with |correlation| ≥ 0.5
   (same as V16 GNN). All features also connect to the GDP target node.
2. **Positional encoding**: Graph-aware positional encoding using node degree
   (number of connections), giving the transformer structural context.
3. **Transformer layers**: Multi-head self-attention with an **attention bias**
   derived from the adjacency matrix — connected nodes get a positive bias,
   encouraging the model to attend to correlated features.
4. **Readout**: The GDP target node's final representation is projected to a
   scalar prediction.

### Fixed Hyperparameters

| Parameter | Value |
|-----------|-------|
| d_model | 32 |
| n_heads | 4 |
| n_layers | 2 |
| dropout | 0.1 |
| epochs | 100 |
| lr | 0.001 |
| batch_size | 16 |
| corr_threshold | 0.5 |

```python
def select_exog_features(train_df: pd.DataFrame,
                         n_features: int = 5) -> List[str]:
    """Select top N features by |correlation| with GDP (training only)."""
    feature_cols = [c for c in train_df.columns if c not in ["GDP"]]
    valid_df = train_df[feature_cols + ["GDP"]].dropna()

    if len(valid_df) < 5:
        return feature_cols[:n_features]

    corrs = valid_df[feature_cols].corrwith(valid_df["GDP"]).abs()
    corrs = corrs.dropna().sort_values(ascending=False)
    return corrs.head(n_features).index.tolist()


def build_edge_index(features: np.ndarray, threshold: float = 0.5):
    """Build graph edges from correlation matrix.

    Edges connect features with |correlation| >= threshold.
    Additionally, all features connect to the last node (GDP target).
    """
    n = features.shape[1]
    corr = np.corrcoef(features.T)

    edges = []
    for i in range(n):
        for j in range(n):
            if i != j and not np.isnan(corr[i, j]) and abs(corr[i, j]) >= threshold:
                edges.append([i, j])

    # Ensure GDP target node (last) is connected to all others
    target_idx = n - 1
    existing = set((e[0], e[1]) for e in edges)
    for i in range(n):
        if i != target_idx:
            if (i, target_idx) not in existing:
                edges.append([i, target_idx])
            if (target_idx, i) not in existing:
                edges.append([target_idx, i])

    if not edges:
        edges = [[0, 1], [1, 0]]

    edge_index = torch.tensor(edges, dtype=torch.long).t().contiguous()
    return edge_index


def edge_index_to_adj(edge_index: torch.Tensor, n_nodes: int) -> torch.Tensor:
    """Convert edge_index to dense adjacency matrix (with self-loops)."""
    adj = torch.zeros(n_nodes, n_nodes)
    if edge_index.numel() > 0:
        adj[edge_index[0], edge_index[1]] = 1.0
    adj += torch.eye(n_nodes)  # self-loops
    return adj


class GraphTransformerLayer(nn.Module):
    """Single Graph Transformer layer with adjacency-biased attention."""

    def __init__(self, d_model: int, n_heads: int, dropout: float = 0.1):
        super().__init__()
        self.d_model = d_model
        self.n_heads = n_heads
        self.head_dim = d_model // n_heads

        self.W_q = nn.Linear(d_model, d_model)
        self.W_k = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)
        self.W_o = nn.Linear(d_model, d_model)

        # Learnable edge bias per head
        self.edge_bias = nn.Parameter(torch.zeros(n_heads))

        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)
        self.ffn = nn.Sequential(
            nn.Linear(d_model, d_model * 4),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(d_model * 4, d_model),
            nn.Dropout(dropout),
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, adj):
        """
        x:   (batch, n_nodes, d_model)
        adj: (batch, n_nodes, n_nodes) — adjacency with self-loops
        """
        B, N, D = x.shape
        H = self.n_heads
        dk = self.head_dim

        # Multi-head attention
        Q = self.W_q(x).view(B, N, H, dk).transpose(1, 2)  # (B, H, N, dk)
        K = self.W_k(x).view(B, N, H, dk).transpose(1, 2)
        V = self.W_v(x).view(B, N, H, dk).transpose(1, 2)

        attn = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(dk)  # (B,H,N,N)

        # Add graph structure bias: connected nodes get a positive bias
        # edge_bias: (H,) → (1, H, 1, 1)
        bias = self.edge_bias.view(1, H, 1, 1) * adj.unsqueeze(1)  # (B,H,N,N)
        attn = attn + bias

        attn = F.softmax(attn, dim=-1)
        attn = self.dropout(attn)

        out = torch.matmul(attn, V)  # (B, H, N, dk)
        out = out.transpose(1, 2).contiguous().view(B, N, D)
        out = self.W_o(out)

        # Residual + LayerNorm
        x = self.norm1(x + self.dropout(out))

        # FFN + Residual + LayerNorm
        x = self.norm2(x + self.ffn(x))

        return x


class GraphTransformerModel(nn.Module):
    """Graph Transformer for GDP level prediction.

    Each node = one feature. Graph edges from correlation.
    Attention is biased by adjacency to encourage attending to
    correlated features. The GDP target node's representation
    is projected to a scalar prediction.
    """

    def __init__(self, input_dim: int, n_nodes: int,
                 d_model: int = 32, n_heads: int = 4,
                 n_layers: int = 2, dropout: float = 0.1):
        super().__init__()
        self.n_nodes = n_nodes

        # Project each node's scalar feature to d_model
        self.input_proj = nn.Linear(input_dim, d_model)

        # Degree-based positional encoding
        self.pos_embed = nn.Embedding(64, d_model)  # max degree 63

        # Transformer layers
        self.layers = nn.ModuleList([
            GraphTransformerLayer(d_model, n_heads, dropout)
            for _ in range(n_layers)
        ])

        # Readout: project GDP target node to scalar
        self.readout = nn.Sequential(
            nn.Linear(d_model, d_model),
            nn.ReLU(),
            nn.Linear(d_model, 1),
        )

    def forward(self, x, adj):
        """
        x:   (batch, n_nodes, input_dim)  — node features
        adj: (batch, n_nodes, n_nodes)    — adjacency matrix
        """
        # Project input features
        h = self.input_proj(x)  # (B, N, d_model)

        # Add positional encoding based on node degree
        degree = adj.sum(dim=-1).long().clamp(max=63)  # (B, N)
        h = h + self.pos_embed(degree)

        # Apply transformer layers
        for layer in self.layers:
            h = layer(h, adj)

        # Readout from GDP target node (last node)
        target_repr = h[:, -1, :]  # (B, d_model)
        return self.readout(target_repr).squeeze(-1)  # (B,)


print("Graph Transformer model defined.")
print(f"  Architecture: {GT_PARAMS['n_layers']}-layer Transformer "
      f"with adjacency-biased attention")
print(f"  d_model={GT_PARAMS['d_model']}, "
      f"n_heads={GT_PARAMS['n_heads']}, "
      f"dropout={GT_PARAMS['dropout']}")
```

## Step 7 — Walk-Forward Evaluation

Same protocol as V16 GNN:
- Walk-forward from 2006-Q1 to 2025-Q4
- Each fold: train on all quarters before test quarter
- Per-fold feature selection, normalization, and graph construction (no leakage)
- Predict GDP level → convert to SAAR for evaluation

```python
def fmt_q(dt: pd.Timestamp) -> str:
    return f"{dt.year}-Q{(dt.month - 1) // 3 + 1}"


def compute_saar(current_gdp: float, prev_gdp: float) -> float:
    """Compute SAAR growth rate from GDP levels."""
    if prev_gdp <= 0:
        return 0.0
    return ((current_gdp / prev_gdp) ** 4 - 1) * 100
```

```python
%%time
# ============================================================
# Graph Transformer Walk-Forward Evaluation (Fixed Hyperparameters)
# Same protocol as V16 GNN
# ============================================================

results_csv = OUTPUT_DIR / "walkforward_results.csv"

log("\n" + "=" * 65)
log("WALK-FORWARD EVALUATION — Graph Transformer Baseline (Month-1 Cutoff)")
log("=" * 65)
log(f"  Exogenous features:  Top {N_EXOG} by |correlation| (per-fold)")
log(f"  Graph Transformer:   d_model={GT_PARAMS['d_model']}, "
    f"heads={GT_PARAMS['n_heads']}, "
    f"layers={GT_PARAMS['n_layers']}, "
    f"epochs={GT_PARAMS['epochs']}")
log(f"  Graph edges:         |corr| >= {CORR_THRESH}")
log(f"  Data cutoff:         Month-1 (same as V16 GNN)")

test_dates = gdp_q.loc[TEST_START:TEST_END].index.tolist()
log(f"  Folds:               {len(test_dates)}")

# Set seed for reproducibility
torch.manual_seed(GT_PARAMS["seed"])
np.random.seed(GT_PARAMS["seed"])

results_list = []

for fold_i, test_q in enumerate(test_dates):
    t0 = time.time()

    # Split: training = all quarters before test_q
    train_mask = quarterly_df.index < test_q
    train_df = quarterly_df[train_mask].dropna(subset=["GDP"])

    if len(train_df) < MIN_TRAIN_Q:
        log(f"  Fold {fold_i:2d} | {fmt_q(test_q)} "
            f"— skipped (only {len(train_df)} train quarters)")
        continue

    # Per-fold feature selection (leakage-free)
    exog_cols = select_exog_features(train_df, N_EXOG)

    # Prepare training features and target
    # Columns: exog_cols + GDP_feature (as nodes), GDP as target
    node_cols = exog_cols  # each column = one graph node
    train_features = train_df[node_cols].copy().ffill().bfill().fillna(0)
    train_target = train_df["GDP"].values.copy()

    # Z-score normalize using training statistics only
    feat_mean = train_features.mean()
    feat_std  = train_features.std().replace(0, 1)
    tgt_mean  = train_target.mean()
    tgt_std   = train_target.std()
    if tgt_std == 0:
        tgt_std = 1.0

    train_features_z = ((train_features - feat_mean) / feat_std).values
    train_target_z   = (train_target - tgt_mean) / tgt_std

    # Build graph from training data correlations
    # Add target as last node → n_nodes = n_features + 1
    train_with_target = np.column_stack([train_features_z, train_target_z])
    n_nodes = train_with_target.shape[1]
    edge_index = build_edge_index(train_with_target, CORR_THRESH)
    adj = edge_index_to_adj(edge_index, n_nodes)

    # Create training samples: each quarter = one graph snapshot
    # Node features: [feature_value] (dim=1 per node)
    X_train_list = []
    y_train_list = []
    for i in range(len(train_features_z)):
        # Node features: exog features + target (but target set to 0 for input)
        node_feats = np.zeros((n_nodes, 1), dtype=np.float32)
        for j in range(len(node_cols)):
            node_feats[j, 0] = train_features_z[i, j]
        # Last node (GDP target): use 0 (model must predict it)
        node_feats[-1, 0] = 0.0
        X_train_list.append(node_feats)
        y_train_list.append(train_target_z[i])

    X_train = np.array(X_train_list)  # (n_quarters, n_nodes, 1)
    y_train = np.array(y_train_list)

    if len(X_train) < 2:
        log(f"  Fold {fold_i:2d} | {fmt_q(test_q)} "
            f"— skipped (too few samples)")
        continue

    # Build and train Graph Transformer
    model = GraphTransformerModel(
        input_dim=1,
        n_nodes=n_nodes,
        d_model=GT_PARAMS["d_model"],
        n_heads=GT_PARAMS["n_heads"],
        n_layers=GT_PARAMS["n_layers"],
        dropout=GT_PARAMS["dropout"],
    ).to(DEVICE)

    # Training
    model.train()
    optimizer = torch.optim.Adam(model.parameters(), lr=GT_PARAMS["lr"])
    criterion = nn.MSELoss()

    X_t = torch.FloatTensor(X_train).to(DEVICE)
    y_t = torch.FloatTensor(y_train).to(DEVICE)
    adj_t = adj.unsqueeze(0).expand(len(X_train), -1, -1).to(DEVICE)

    batch_size = GT_PARAMS["batch_size"]
    n_samples = len(X_t)

    for epoch in range(GT_PARAMS["epochs"]):
        indices = torch.randperm(n_samples)
        for start in range(0, n_samples, batch_size):
            end = min(start + batch_size, n_samples)
            idx = indices[start:end]

            batch_X = X_t[idx]
            batch_y = y_t[idx]
            batch_adj = adj_t[idx]

            optimizer.zero_grad()
            pred = model(batch_X, batch_adj)
            loss = criterion(pred, batch_y)
            loss.backward()
            optimizer.step()

    # Predict test quarter
    if test_q not in quarterly_df.index:
        continue
    test_exog = quarterly_df.loc[[test_q], node_cols].copy()
    test_exog = test_exog.ffill().bfill().fillna(0)
    test_features_z = ((test_exog - feat_mean[node_cols]) / feat_std[node_cols]).values

    # Build test graph input
    test_node_feats = np.zeros((1, n_nodes, 1), dtype=np.float32)
    for j in range(len(node_cols)):
        test_node_feats[0, j, 0] = test_features_z[0, j]
    test_node_feats[0, -1, 0] = 0.0  # GDP target node = 0

    test_X = torch.FloatTensor(test_node_feats).to(DEVICE)
    test_adj = adj.unsqueeze(0).to(DEVICE)

    model.eval()
    with torch.no_grad():
        pred_z = model(test_X, test_adj).item()

    # De-normalize
    pred_gdp = pred_z * tgt_std + tgt_mean

    # Compute SAAR
    actual_gdp = float(gdp_q.loc[test_q, "GDP"])
    prev_q_dt  = test_q - pd.DateOffset(months=3)
    if prev_q_dt in gdp_q.index:
        prev_gdp = float(gdp_q.loc[prev_q_dt, "GDP"])
    else:
        prev_qs  = gdp_q.index[gdp_q.index < test_q]
        prev_gdp = float(gdp_q.loc[prev_qs[-1], "GDP"])

    actual_saar = compute_saar(actual_gdp, prev_gdp)
    pred_saar   = compute_saar(pred_gdp, prev_gdp)
    saar_error  = pred_saar - actual_saar
    saar_ae     = abs(saar_error)
    dir_ok      = (pred_saar >= 0) == (actual_saar >= 0)

    elapsed = time.time() - t0
    mark = "\u2713" if dir_ok else "\u2717"

    log(f"  Fold {fold_i:2d} | {fmt_q(test_q)} | "
        f"Act={actual_saar:+7.2f}  Pred={pred_saar:+7.2f} | "
        f"AE={saar_ae:5.2f} | {mark} | {elapsed:.1f}s")

    n_edges = edge_index.shape[1] if edge_index.numel() > 0 else 0
    row = {
        "fold": fold_i,
        "test_quarter": test_q.strftime("%Y-%m-%d"),
        "actual_gdp": actual_gdp,
        "pred_gdp": round(pred_gdp, 2),
        "prev_gdp": prev_gdp,
        "actual_saar": round(actual_saar, 4),
        "pred_saar": round(pred_saar, 4),
        "saar_error": round(saar_error, 4),
        "saar_abs_error": round(saar_ae, 4),
        "direction_correct": bool(dir_ok),
        "n_train": len(train_df),
        "exog_features": ", ".join(exog_cols),
        "n_features": len(exog_cols),
        "n_nodes": n_nodes,
        "n_edges": n_edges,
        "elapsed_sec": round(elapsed, 2),
    }
    results_list.append(row)

    # Running stats
    running_mae = np.mean([r["saar_abs_error"] for r in results_list])
    running_dir = sum(1 for r in results_list if r["direction_correct"])
    log(f"         Running MAE: {running_mae:.3f} | "
        f"Dir: {running_dir}/{len(results_list)}")

# ── Save results ─────────────────────────────────────────────
rdf = pd.DataFrame(results_list)
rdf.to_csv(results_csv, index=False)

log(f"\n{'=' * 65}")
log(f"WALK-FORWARD COMPLETE — {len(results_list)} folds")
log(f"{'=' * 65}")

final_mae = rdf["saar_abs_error"].mean()
final_dir = rdf["direction_correct"].sum()
log(f"  Final MAE:       {final_mae:.3f} pp")
log(f"  Final Direction: {int(final_dir)}/{len(rdf)}")

# Save summary JSON
summary = {
    "model": "GraphTransformer_Month1_Baseline",
    "gt_params": GT_PARAMS,
    "corr_threshold": CORR_THRESH,
    "data_cutoff": "month-1 (Jan/Apr/Jul/Oct)",
    "timing_advantage": "~45 days before GDP advance estimate",
    "n_folds": len(rdf),
    "n_exog": N_EXOG,
    "saar_mae": round(float(final_mae), 4),
    "direction_accuracy": f"{int(final_dir)}/{len(rdf)}",
}
with open(OUTPUT_DIR / "summary.json", "w") as f:
    json.dump(summary, f, indent=2)
log(f"  Saved to {OUTPUT_DIR}")
```

### 9g. Per-Quarter Detail with Prediction "As-Of" Date

For each quarter, we compute the **prediction as-of date** (when all
month-1 data is available, accounting for MRTS ~42-day publication lag)
and how many **days ahead** of the BEA GDP advance estimate this
prediction is made.

```python
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', 180)

detail = results.copy()


def compute_as_of_date(q):
    """Month-1 end date + 42 days MRTS publication lag."""
    month1 = pd.Timestamp(year=q.year, month=q.month, day=1)
    month1_end = month1 + pd.offsets.MonthEnd(0)
    return month1_end + pd.Timedelta(days=42)


def compute_gdp_release(q):
    """GDP advance estimate: ~28 days after quarter end."""
    qtr_end_month = {1: 3, 4: 6, 7: 9, 10: 12}[q.month]
    qtr_end = (pd.Timestamp(year=q.year, month=qtr_end_month, day=1)
               + pd.offsets.MonthEnd(0))
    return qtr_end + pd.Timedelta(days=28)


detail["as_of_date"]  = detail["test_quarter"].apply(compute_as_of_date)
detail["gdp_release"] = detail["test_quarter"].apply(compute_gdp_release)
detail["days_ahead"]  = (detail["gdp_release"]
                         - detail["as_of_date"]).dt.days

display_df = detail[[
    'test_quarter', 'as_of_date', 'gdp_release', 'days_ahead',
    'actual_saar', 'pred_saar', 'saar_abs_error', 'direction_correct',
]].copy()

display_df['Quarter']     = display_df['test_quarter'].apply(fmt_q)
display_df['As-Of']       = display_df['as_of_date'].dt.strftime('%Y-%m-%d')
display_df['GDP Release'] = display_df['gdp_release'].dt.strftime('%Y-%m-%d')
display_df['Days Ahead']  = display_df['days_ahead']
display_df['Actual']      = display_df['actual_saar'].round(2)
display_df['Predicted']   = display_df['pred_saar'].round(2)
display_df['Abs Err']     = display_df['saar_abs_error'].round(2)
display_df['Dir']         = display_df['direction_correct'].map(
    {True: '\u2713', False: '\u2717'})

print_df = display_df[[
    'Quarter', 'As-Of', 'GDP Release', 'Days Ahead',
    'Actual', 'Predicted', 'Abs Err', 'Dir',
]]

print(print_df.to_string(index=False))
print(f"\nAverage days ahead of GDP release: "
      f"{detail['days_ahead'].mean():.0f} days")
```

### 9j. GFC Negative Quarters

Quarters during the 2008-2009 financial crisis where GDP growth was

## Methodology Notes

### Walk-Forward Protocol (identical to V16 GNN)
- **No future data leakage**: Each fold trains only on data before the test quarter
- **Per-fold feature selection**: Top 5 features by |correlation| on training data only
- **Per-fold normalization**: Z-score using training statistics only
- **Per-fold graph construction**: Edges from training correlation only
- **Month-1 cutoff**: Only first month of each quarter used

### Graph Transformer Specifics
- **Architecture**: 2-layer Transformer with adjacency-biased multi-head attention
- **Graph construction**: Correlation-based edges (|corr| >= 0.5) + GDP hub connections
- **Positional encoding**: Node degree embedding (structural, not sequential)
- **Attention bias**: Adjacency matrix modulates attention scores — connected nodes
  (correlated features) receive a learnable positive bias
- **Fixed hyperparameters**: d_model=32, n_heads=4, epochs=100, lr=0.001
- **No tuning**: No grid search, no cross-validation, no hyperparameter optimization
- **Node features**: Each node = one economic indicator (scalar value per quarter)
- **GDP target node**: Last node, connected to all others, input set to 0
- **Readout**: Target node representation → MLP → scalar GDP prediction
- **No ensemble**: Single model per fold (unlike GNN's 9-member trimmed mean ensemble)

### Key Differences from V16 GNN
- **Attention mechanism**: Learned feature weighting vs fixed message passing
- **No lookback window**: Single quarter per sample (vs GNN's monthly lookback)
- **Simpler features**: 5 per-fold selected (vs GNN's 39 fixed features)
- **No ensemble**: Single model (vs GNN's trimmed mean of 9)
- **No edge dropout**: No data augmentation during training

### Evaluation Periods (identical to V16 GNN)
- **Overall**: All 80 folds (2006-Q1 to 2025-Q4)
- **Recession quarters**: 2008-Q3, 2008-Q4, 2009-Q1, 2020-Q2, 2020-Q3
- **GFC period**: 2008-Q1 through 2009-Q4 (8 quarters)
- **GFC negative**: 2008-Q1, 2008-Q4, 2009-Q1, 2009-Q2 (contraction quarters)
- **COVID period**: 2020-Q1 through 2021-Q2 (6 quarters)
- **COVID negative**: 2020-Q1, 2020-Q2 only
- **Post-2011-Q3**: GDPNow comparison period
- **Eras**: Pre-GFC, GFC, Recovery, Expansion, COVID, Post-COVID
